```python
# import libraries
import csv
import numpy as np 
import pandas as pd 
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
# load dataset
df = pd.read_csv('dataDir/mb_full.csv')
print(df.shape)
df.tail()
```

```python
df = df[['first_seen_utc', 'signature', 'tlsh', 'ssdeep']]
df.tail()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset = ["signature"], inplace=True)
df.reset_index(drop=True, inplace=True)
df.isnull().sum()
```

```python
print(df.shape)
df.tail()
```

# Same Name For Malware

```python
df["signature"] = df["signature"].str.lower()
```

```python
df["signature"] = df["signature"].str.replace(" ","")
df["signature"] = df["signature"].str.replace("-","")
df["signature"] = df["signature"].str.replace("_","")
df["signature"] = df["signature"].str.replace("ransomware.","")
df["signature"] = df["signature"].str.replace("adware.","")
df["signature"] = df["signature"].str.replace("downloader.","")
df["signature"] = df["signature"].str.replace("spambot.","")
df["signature"] = df["signature"].str.replace("worm.","")
df["signature"] = df["signature"].str.replace("coinminer.","")
df["signature"] = df["signature"].str.replace("riskware.","")
df["signature"] = df["signature"].str.replace("asprox.","")
```

```python
#replace malware with same name
df['signature'] = df['signature'].replace(['mirai','gafgyt'],'mirai/gafgyt')
df[df['signature'].str.contains("mirai/gafgyt")]
```

```python
df.signature.nunique()
df.signature.unique()
```
