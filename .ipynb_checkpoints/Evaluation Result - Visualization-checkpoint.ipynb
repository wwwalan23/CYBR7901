```python
from sklearn.cluster import KMeans
from sklearn import metrics
import time

srange=[10,25,100,150]

ac_time = [41.4756/60,267.7342/60]
dbscan_time = [67.2022/60,425.5904/60,6120.2715/60, 12386.7338/60]
hact_time = [6.1518/60,31.8061/60,462.4878/60, 845.5256/60]
optics_time = [109.9218/60,695.648/60,10623.5125/60, 22925.9930/60]

ac_time_ssdeep = [80.8111/60,466.7641/60]
dbscan_time_ssdeep = [145.3149/60,895.5755/60,14498.7721/60]
optics_time_ssdeep = [226.9109/60,1392.4383/60,22342.3134/60]

import matplotlib.pyplot as plt
line1, = plt.plot([10,25], ac_time,label='TLSH_AC')
line2, = plt.plot(srange, dbscan_time,label='TLSH_DBSCAN')
line3, = plt.plot(srange, hact_time,label='TLSH_HAC-T')
line4, = plt.plot(srange, optics_time,label='TLSH_OPTICS')

line1s, = plt.plot([10,25], ac_time_ssdeep,label='SSDEEP_AC', linestyle="--")
line2s, = plt.plot([10,25,100], dbscan_time_ssdeep,label='SSDEEP_DBSCAN', linestyle="--")
line4s, = plt.plot([10,25,100], optics_time_ssdeep,label='SSDEEP_OPTICS', linestyle="--")

plt.legend(handles=[line1, line2, line3, line4, line1s, line2s, line4s], loc='upper left')

plt.xlabel('Sample size (K)', fontsize=15)
plt.ylabel('Time (min)', fontsize=15)
plt.show()
```
